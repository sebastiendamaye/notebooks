```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Load the trained model
model = tf.keras.models.load_model('dga_detection_model_local.keras')

# Check the model summary to confirm it's loaded correctly
model.summary()
```

```python
# Read the file with domain names
with open('test_sample.txt', 'r') as f:
    domains = [line.strip() for line in f.readlines()]

# Assume tokenizer is already fitted with the data during training
tokenizer = Tokenizer()
tokenizer.fit_on_texts(domains)

# Tokenize the domains
sequences = tokenizer.texts_to_sequences(domains)

# Pad the sequences to ensure they are the same length
padded_domains = pad_sequences(sequences, maxlen=75)  # Adjust 'maxlen' based on your training

# Get predictions
predictions = model.predict(padded_domains)

# Convert predictions to binary labels (if needed)
labels = (predictions > 0.5).astype(int)

# Print predictions
for domain, prediction in zip(domains, labels):
    print(f"Domain: {domain} | Prediction: {'Malware' if prediction == 1 else 'Legitimate'}")
```
